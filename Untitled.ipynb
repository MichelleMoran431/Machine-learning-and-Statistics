```python
import pandas as pd

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category = FutureWarning)
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category = FutureWarning)



heart = pd.read_csv(r'C:\Users\User\Desktop\repo\Machine-learning-and-Statistics\heart.csv')


heart.columns = ['age', 'sex', 'cp', 'trestbps', 'chol',
              'fbs', 'restecg', 'thalach', 'exang', 
              'oldpeak', 'slope', 'HeartDisease']

### 1 = male, 0 = female
heart.isnull().sum()

#heart['HeartDisease'] = heart.HeartDisease.map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
heart['sex'] = heart.sex.map({0: 'female', 1: 'male'})


import matplotlib.pyplot as plt
import seaborn as sns

# distribution of target vs age 
sns.set_context("paper", font_scale = 1, rc = {"font.size": 10,"axes.titlesize": 10,"axes.labelsize": 10}) 
sns.catplot(kind = 'count', data = heart, x = 'age', hue = 'HeartDisease', order = heart['age'].sort_values().unique())
plt.title('Variation of Age for each target class')
plt.show()

 
# barplot of age vs sex with hue = target
sns.catplot( data = heart, y = 'age', x = 'sex', hue = 'HeartDisease')
plt.title('Distribution of age vs sex with the target class')
plt.show()

heart['sex'] = heart.sex.map({'female': 0, 'male': 1})


```
